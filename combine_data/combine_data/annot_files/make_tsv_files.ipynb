```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
annots = pd.read_table('3_5_2_6_annotations.txt')
```

```python
annots
```

```python
names = ""

with open('../nwk_files/3_5_2_6_filt.nwk', 'r') as nwk:
    for line in nwk:
        names += line
        
```

```python
first = names.split(',')

tester = [x.split(':')[0] for x in first]

stripped = [x.strip('((') for x in tester]

len(stripped)
```

```python

tsv = pd.DataFrame()
annot_names = []
annot_data = []


counter = defaultdict(int)
filtered = annots.loc[annots['Entry'].isin(stripped)]
hit = ['cephalothin', 'cephalotin', 'cefalotin']
km_data = filtered[['Entry', 'BRENDA_KM']]

for entry, x in zip(km_data['Entry'], km_data['BRENDA_KM']):
    
    if x != 'None':
        thing = x.split(';')
        i = 0
        while i < len(thing) - 1:
           
            holder = thing[i:i+2]
            name = holder[1].split('_')[0].lower()
            if name in hit:
                print(entry)
                annot_names.append(entry)
                annot_data.append(holder[0])
            counter[name] += 1
            i += 2


tsv['Entry'] = annot_names
tsv['km'] = annot_data

tsv
```

```python
#remove double up of entry 11

tsv.drop(tsv.index[11], axis=0, inplace=True)
tsv
```

```python
tsv.to_csv('3_5_2_6_km_class_A', sep='\t')
```
